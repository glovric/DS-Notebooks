# Data preprocessing

```python
# Switching from pd.DataFrame to np.ndarray
X = df[df.columns[:-1]].to_numpy(dtype=np.float64)
y = df[df.columns[-1]].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Normalize by scaling to mean 0 and variance 1
X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

# Model selection

```python
import torch
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm

class MagicDataset(Dataset):

    def __init__(self, data, labels):
        self.data = data
        self.labels = labels

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = self.data[idx]
        sample_label = self.labels[idx]
        return sample, sample_label
    
class NeuralNet:

    def __init__(self, batch_size=100, epochs=5, learning_rate=1e-3, layers=[torch.nn.Linear(10, 64),
                                                                             torch.nn.ReLU(),
                                                                             torch.nn.Dropout(p=0.2),
                                                                             torch.nn.Linear(64, 64),
                                                                             torch.nn.ReLU(),
                                                                             torch.nn.Dropout(p=0.2),
                                                                             torch.nn.Linear(64, 1),
                                                                             torch.nn.Sigmoid()
                                                                            ]):

        self.model = torch.nn.Sequential(*layers)

        self.loss_fn = torch.nn.BCELoss()
        self.optimizer = torch.optim.Adam(self.model.parameters(), learning_rate)

        self.batch_size = batch_size
        self.epochs = epochs

    def fit(self, X, y):

        X, y = torch.tensor(X).float(), torch.tensor(y).float().unsqueeze(-1)

        trainset = MagicDataset(data=X, labels=y)
        train_loader = DataLoader(trainset, batch_size=self.batch_size, shuffle=True, num_workers=2, pin_memory=True)

        for e in range(self.epochs):
                    
            loop = tqdm(enumerate(train_loader), total=len(train_loader), leave=True)

            for idx, (x, y_true) in loop:

                self.optimizer.zero_grad()
                y_pred = self.model(x)
                loss = self.loss_fn(y_pred, y_true)
                loss.backward()
                self.optimizer.step()

                accuracy = (y_pred.round() == y_true).float().mean()
                accuracy = float(accuracy)

                loop.set_description(f"Epoch [{e+1}/{self.epochs}")
                loop.set_postfix(loss=loss.item(), acc=accuracy)

    def predict(self, X):

        X = torch.tensor(X).float()
        return (self.model(X).detach().numpy() >= 0.5).astype(int)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, Perceptron, RidgeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

models = {'SVM': SVC(kernel='rbf'),
          'LogReg': LogisticRegression(),
          'Perceptron': Perceptron(),
          'Ridge': RidgeClassifier(),
          'KNN': KNeighborsClassifier(),
          'RandomForest': RandomForestClassifier(),
          'AdaBoost': AdaBoostClassifier(algorithm='SAMME'),
          'GradBoost': GradientBoostingClassifier(),
          'HistGradBoost': HistGradientBoostingClassifier(),
          'NaiveBayes': GaussianNB(),
          'DecisionTree': DecisionTreeClassifier(),
          'NeuralNet': NeuralNet()
          }
```

# Model training

```python
from sklearn.metrics import accuracy_score

train_results = []

for m in models:
    models[m].fit(X_train, y_train)
    y_pred = models[m].predict(X_test)
    # Store model name and model accuracy
    train_results.append((m, accuracy_score(y_test, y_pred)))

# Sort the training results by highest accuracy
train_results.sort(key=lambda x: x[1], reverse=True)

for model_name, model_accuracy in train_results:
    print(f'{model_name} accuracy: {model_accuracy * 100:.2f}%')
```

# Selecting the best performing models based on accuracy

```python
best_models = {}

for model_name, model_accuracy in train_results[:4]:
    best_models[model_name] = models[model_name]
```

# Reports and confusion matrices

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

for m in best_models:
    y_pred = best_models[m].predict(X_test)
    print(classification_report(y_test, y_pred))
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(m)
    plt.show()
```

# Grid search - hyperparameter tuning

```python
max_iter = [100, 200, 300]
learning_rate = [0.01, 0.1, 0.5]
max_bins = [100, 200, 255]

best_acc = 0

for m in max_iter:
    for lr in learning_rate:
        for b in max_bins:
            model = HistGradientBoostingClassifier(max_iter=m, learning_rate=lr, max_bins=b)
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            acc = accuracy_score(y_test, y_pred)
            if acc > best_acc:
                best_acc = acc
                best_models['HistGradBoost'] = model
```

```python
n_estimators = [50, 100, 200]
criterion = ['gini', 'entropy', 'log_loss']

best_acc = 0

for n in n_estimators:
    for c in criterion:
        model = RandomForestClassifier(n_estimators=n, criterion=c)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        if acc > best_acc:
            best_acc = acc
            best_models['RandomForest'] = model
```

```python
n_estimators = [50, 100, 200]
learning_rate = [0.01, 0.1, 0.5]

best_acc = 0

for n in n_estimators:
    for lr in learning_rate:
        model = GradientBoostingClassifier(n_estimators=n, learning_rate=lr)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        if acc > best_acc:
            best_acc = acc
            best_models['GradBoost'] = model
```

```python
c = [0.1, 1, 10]
kernel = ['linear', 'rbf', 'poly']

best_acc = 0

for c in c:
    for k in kernel:
        model = SVC(C=c, kernel=k)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        if acc > best_acc:
            best_acc = acc
            best_models['SVM'] = model
        
```
